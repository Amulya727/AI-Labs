```python
# importing collections
from collections import deque

class Gra:
   

    def __init__(self, list):
        self.list = list

    def get_neighbors(self, x):
        return self.list[x]

    # estimated time (heuristic function)  for all graph nodes
    def h(self, node):
        H = {
            #bee start node
            'B': 2,
             #emptyland1 node
            'E1': 4,
             #emptyland2 node
            'E2': 2,
             #FlowersLand(goal node)
            'F': 0
        }

        return H[node]

    def a_Star(self, startNode, stopNode):
        # open_list is a list of landNodes which have been visited, but who's neighbors
        # haven't all been checked on , starts off with the startNode
        # closed_list is a list of landNodes which have been visited
        # and who's neighbors have been checked on
        open_list = set([startNode])
        closed_list = set([])

        # g contains current cost from startNode to all other nodes
       
        g = {}

        g[startNode] = 0

        # previous contains an adjacency map of all nodes
        previous = {}
        previous[startNode] = startNode

        while len(open_list) > 0:
            node = None

            # find a node with the lowest value of f() - evaluation function
            for x in open_list:
                if node == None or g[x] + self.h(x) < g[node] + self.h(node):
                    node = x;

            if node == None:
                print('ThisWayTakesMoreTime!')
                return None

            # if the current node  is the stopCity
            # then we begin reconstructing the path from it to the startNode
            if node == stopNode:
                reconst_path = []

                while previous[node] != node:
                    reconst_path.append(node)
                    node = previous[node]

                reconst_path.append(startNode)

                reconst_path.reverse()

                #print('LeastDistanceToFlowerLand: {}'.format(reconst_path))
                return reconst_path

            # for all neighborlandnodes of the current node do
            for (m, weight) in self.get_neighbors(node):
                # if the current node isn't in both open_list and closed_list
                # add it to open_list and note node as it's previous
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    previous[m] = node
                    g[m] = g[node] + weight

                # otherwise, check if it's quicker to first visit node, then m
                # and if it is, update previous data and g data
                # and if the node was in the closed_list, move it to open_list
                else:
                    if g[m] > g[node] + weight:
                        g[m] = g[node] + weight
                        previous[m] = node

                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            # remove node from the open_list, and add it to closed_list
            # because all of his neighbournodes were inspected
            open_list.remove(node)
            closed_list.add(node)

        print('ThisWayTakesMoreTime!')
        return None
```

```python
list = {
    'B': [('E1', 9), ('E2', 20), ('F', 30)],
    'E1': [('F', 20)],
    'E2': [('F', 10)]
}
graph1 = Gra(list)
graph1.a_Star('B', 'F')
```
